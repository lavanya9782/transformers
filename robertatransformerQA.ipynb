```python
import torch
from transformers import AutoModelForQuestionAnswering,AutoTokenizer,pipeline
```

```python
QA_input = [{
    'question': 'Why is model conversion important?',
    'context': 'The option to convert models between FARM and transformers gives freedom to the user and let people easily switch between frameworks.'
},
            {
    'question': 'What tablet for fever',
    'context': 'only a physician can prescribe the medicine for fever, the preferred ones are tylenol or paracetamol'
},
            {
    'question': 'what are the  initial symptons for viral fever',
    'context': 'generally a patient with high fever have  high temperature and body ache and followed by cold and cough which last for 3 to 7 days '
}]
```

```python
selected_model_name="deepset/roberta-base-squad2"
```

```python
mymodel=AutoModelForQuestionAnswering.from_pretrained(selected_model_name)
mytokenizer=AutoTokenizer.from_pretrained(selected_model_name)
```

```python
input0=mytokenizer(QA_input[0]['question'],QA_input[0]['context'],return_tensors="pt")
output0=mymodel(**input0)
```

```python
input1=mytokenizer(QA_input[1]['question'],QA_input[1]['context'],return_tensors="pt")
output1=mymodel(**input1)
```

```python
input2=mytokenizer(QA_input[2]['question'],QA_input[2]['context'],return_tensors="pt")
output2=mymodel(**input2)
```

```python
output2
```

```python
answerstartindex=torch.argmax(output0.start_logits)
answerendindex=torch.argmax(output0.end_logits)

answertokens=input0.input_ids[0,answerstartindex:answerendindex+1]
answer=mytokenizer.decode(answertokens)
print("question:{} \n answer{}".format(QA_input[0]['question'],answer))
```

```python
answerstartindex=torch.argmax(output1.start_logits)
answerendindex=torch.argmax(output1.end_logits)

answertokens=input1.input_ids[0,answerstartindex:answerendindex+1]
answer=mytokenizer.decode(answertokens)
print("question:{} \n answer{}".format(QA_input[1]['question'],answer))
```

```python
answerstartindex=torch.argmax(output2.start_logits)
answerendindex=torch.argmax(output2.end_logits)

answertokens=input2.input_ids[0,answerstartindex:answerendindex+1]
answer=mytokenizer.decode(answertokens)
print("question:{} \n answer{}".format(QA_input[2]['question'],answer))
```
